```python
parameters = {
              # 'device':'gpu',
              'max_depth': [ 3, 5, 7, 9],

              'num_leaves':[31, 63, 127],

              # 'min_data_in_leaf':[20],
              # 'min_sum_hessian_in_leaf':[1e-3],
              
              # 'feature_fraction': [0.6],
              # 'bagging_fraction': [1.0],
              # # # 'bagging_freq': [2, 4, 5, 6, 8],
              # 'lambda_l1': [0.8],
              # 'lambda_l2': [30],

              # 'learning_rate': [0.2],
              # # 'cat_smooth': [10, 15, 20, ]
}

best_lgbm = None
best_acc = 0
f1, f2 = np.meshgrid(parameters['max_depth'],parameters['num_leaves'])
f1 = f1.flat[:]
f2 = f2.flat[:]

for i, j in zip(f1, f2):
  lgbm = lgb.LGBMClassifier(objective = 'muilti_class',
                          metric = 'multi_logloss',
                          max_depth = i,
                          num_leaves = j,
                          learning_rate = 0.1,
                          feature_fraction = 1,
                          min_child_samples=21,
                          min_child_weight=0.001,
                          bagging_fraction = 1,
                          bagging_freq = 2,
                          reg_alpha = 0.001,
                          reg_lambda = 8,
                          cat_smooth = 0, 
                          device = 'gpu',
                          gpu_platform_id = 0, 
                          gpu_device_id = 0
                          )
  lgbm.fit(X_train, y_train)
  
  y_test_pred = torch.tensor(lgbm.predict(X_test))
  y_train_pred = torch.tensor(lgbm.predict(X_train))
  acc = sum(y_test_pred == y_test).item()/y_test.size(0)
  acc_train = sum(y_train_pred == y_train).item()/y_train.size(0)
  print(f'{i}, {j}, acc in dev:{acc}, acc in train:{acc_train}')
  if best_acc <= acc:
    best_lgbm = lgbm
    best_acc = acc
  
print(best_lgbm, acc) 
```
